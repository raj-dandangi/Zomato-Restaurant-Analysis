```python
#performing statistical methods T-test: Online Delivery vs. Rating
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols

rating_online_delivery = df[df['has_online_delivery'] == True]['rating']
rating_no_online_delivery = df[df['has_online_delivery'] == False]['rating']

if len(rating_online_delivery) > 1 and len (rating_no_online_delivery) >1 :
    t_stat, p_value = stats.ttest_ind(rating_online_delivery, rating_no_online_delivery, equal_var=False) # Welch's t-test assuming unequal variances
    print(f'mean of rating online delivery : {rating_online_delivery.mean():.2f}')
    print(f'mean of rating no online delivery : {rating_no_online_delivery.mean():.2f}')
    print(f"T-statistic: {t_stat:.3f}")
    print(f"P-value: {p_value:.3f}")
    alpha=0.05
    if p_value < alpha:
        print(f"Conclusion: Reject the Null Hypothesis. There is a statistically significant difference in average ratings between restaurants with and without online delivery (p < {alpha}).")
    else:
        print(f"Conclusion: Fail to Reject the Null Hypothesis. There is no statistically significant difference in average ratings between restaurants with and without online delivery (p >= {alpha}).")
else:
    print("Not enough data in one or both groups to perform t-test.")
    
```

```python
# Perform one-way ANOVA
df['price_range_cat'] = df['price_range'].astype('category')

model = ols('rating ~ C(price_range_cat)', data=df).fit() # C() denotes categorical variable
anova_table = sm.stats.anova_lm(model, typ=2) # typ=2 for Type II sum of squares
print(anova_table)
p_value_anova = anova_table['PR(>F)'][0]
alpha = 0.05
if p_value_anova < alpha:
    print(f"Conclusion: Reject the Null Hypothesis. There is a statistically significant difference in average ratings across different price range categories (p < {alpha}).")
else:
    print(f"Conclusion: Fail to Reject the Null Hypothesis. There is no statistically significant difference in average ratings across different price range categories (p >= {alpha}).")
```

```python
cocaine_columns = ['average_cost_for_two', 'menu_price', 'votes']
col_data = df[cocaine_columns].dropna()
col_data.head(10)
```

```python
for col in cocaine_columns:
    print(f"\n### Column: '{col}' ###")

    # Drop NaNs temporarily for value checks to avoid errors
    col_data = df[col].dropna()

    if col_data.empty:
        print(f"  - WARNING: Column '{col}' is empty after dropping NaNs. Cannot analyze.")
        continue

    # Check for negative values
    has_negatives = (col_data < 0).any()
    print(f"  - Contains Negative Values: {has_negatives}")

    # Check for zero values and count them
    zero_count = (col_data == 0).sum()
    has_zeros = zero_count > 0
    print(f"  - Contains Zero Values: {has_zeros} (Count: {zero_count})")

    # Get min and max values for range
    min_val = col_data.min()
    max_val = col_data.max()
    print(f"  - Value Range: [{min_val:.2f}, {max_val:.2f}]")

    # Suggest transformation based on findings
    print("  - Transformation Recommendation:")
    if has_negatives:
        print("    -> Yeo-Johnson PowerTransformer: Handles negative, zero, and positive values. Most robust.")
    elif has_zeros and zero_count > 0.05 * len(col_data): # More than 5% zeros suggests Yeo-Johnson over log1p
        print("    -> Yeo-Johnson PowerTransformer: Handles zeros well, good for distributions with many zeros.")
        print("    -> log1p (np.log1p): Also handles zeros, but Yeo-Johnson might be more optimal.")
    elif has_zeros and zero_count <= 0.05 * len(col_data): # Few zeros, log1p is fine
        print("    -> log1p (np.log1p): Simple and effective for positive data with some zeros.")
        print("    -> Box-Cox (PowerTransformer, method='box-cox') if NO zeros, otherwise Yeo-Johnson or log1p.")
    else: # Only positive values, no zeros
        print("    -> Box-Cox PowerTransformer (method='box-cox'): Excellent for positive-only data.")
        print("    -> log (np.log): If Box-Cox is overkill and you want a strong compression.")
        print("    -> Square Root (np.sqrt): For milder skewness.")

    print("-" * 40)

print("\n--- Summary of Recommendations ---")
print("1. If a column has negative values, your ONLY robust choice is `Yeo-Johnson PowerTransformer`.")
print("2. If a column has zeros, `Yeo-Johnson PowerTransformer` or `log1p` are good. `Yeo-Johnson` is generally more optimal for more flexibility.")
print("3. If a column has ONLY positive values (no zeros), `Box-Cox PowerTransformer` is often the best for optimal normalization. `log` (np.log) or `sqrt` are simpler alternatives.")

```

# using random forest to find the answer which is the most infulancal factor for voting and popularty

```python
X = df_v2.drop(['rating','votes'],axis=1)

y_rating = df_v2['rating']
cat_cols = ['city','cuisine']
num_feats = ['price_range','average_cost_for_two','has_online_delivery','has_table_booking','menu_price']

pre = ColumnTransformer([

    ('cat',OneHotEncoder(handle_unknown='ignore'),cat_cols),
    ('num','passthrough',num_feats)
])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

rf_rating = Pipeline([('pre',pre),('rf',RandomForestRegressor(n_estimators=200, random_state=0, n_jobs=-1))])
print('r2 rating')
rf_rating.fit(X, y_rating)
print(r2_score(y_rating, rf_rating.predict(X)))
```

```python
imp = rf_rating.named_steps['rf'].feature_importances_
feat_names = rf_rating.named_steps['pre'].get_feature_names_out()
imp_df = pd.DataFrame({'feature':feat_names,'importance':imp}).sort_values('importance', ascending=False).head(12)
print('imp_rating')
print(imp_df)
```
